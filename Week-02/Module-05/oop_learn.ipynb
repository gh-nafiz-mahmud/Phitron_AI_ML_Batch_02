```python
# class and objects

class Phone:
  # attributes
  category = "Electronics"

  # constructor
  # instance er attributes
  def __init__(self, model, battery, camera, battery_percentage):
    self.model = model
    self.battery = battery
    self.camera = camera
    self.battery_percentage = battery_percentage

  # methods
  def charge(self, hour):
    self.charge += hour

  def capture(self, total_photo):
    if(self.battery_percentage) <= 0:
      print("No charge")
    else:
      self.battery_percentage -= total_photo
      print(f"Photo captured in {self.model}")

# creating object
apple = Phone("Iphone 17", 4000, 40, 100)
apple.category = "Super electronics"
apple.capture(1)
# blueberry = Phone("B-17", 5000, 30)
# motorolaa = Phone("M-17", 5500, 40)

# print(type(apple))
# print(apple.category, apple.camera, apple.battery, apple.battery_percentage)
print(apple.battery_percentage)

# print(motorolaa.category)
# print(blueberry.category)
```
